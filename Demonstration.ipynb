## Doing QUERIES


```
df = splusdata.queryidr3('main3.1', ['RA > 48.9', 'RA < 49'], columns=['ID', 'RA', 'DEC'])
```

```
df = splusdata.queryidr3('stripe82', conditions = ['FWHM > 80'], password = '##')
```

You may also want to do complex queries


```
df = splusdata.queryidr3(Survey = 'main3.1', conditions = ['(R_PStotal - F660_PStotal) > 0.6', '(I_PStotal - F660_PStotal) > 0.6', 'R_PStotal > 0', 'R_PStotal < 23', 'F515_PStotal > 0', 'F515_PStotal < 23', 'F861_PStotal > 0', 'F861_PStotal < 23', 'I_PStotal > 0', 'I_PStotal < 23', 'R_PStotal < 20.5'], columns='all')
```
